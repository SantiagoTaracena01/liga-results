```python
# Proceso de importar pandas para obtener los resultados anteriores.
import pandas as pd
```

```python
# Lectura del .csv con los resultados obtenidos.
data = pd.read_csv("./results/results.csv")
data.index = range(1, (len(data) + 1))
data
```

```python
# Proceso para importar matplotlib y graficar los resultados.
import matplotlib.pyplot as plt
```

```python
# Gráfico para observar la distribución de puntos de los equipos.
plt.figure(figsize=(7, 5))
plt.barh(data["Team"], data["Pts."], color="skyblue")
plt.xlabel("Puntos")
plt.ylabel("Equipo")
plt.title("Puntos de los Equipos en La Liga 2023-24")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Gráfico para observar los goles a favor y en contra de cada equipo.
data[["GF", "GA"]].plot(kind="bar", stacked=True, figsize=(9, 5))
plt.xlabel("Equipo")
plt.ylabel("Goles")
plt.title("Goles a Favor (GF) y en Contra (GA) de los Equipos en La Liga 2023-24")
plt.xticks(range(len(data["Team"])), data["Team"], rotation=45)
plt.legend(["Goles a Favor (GF)", "Goles en Contra (GA)"])
plt.show()
```

```python
# Gráfico para observar las victorias, empates y derrotas de cada equipo.
data[["W", "D", "L"]].plot(kind="bar", figsize=(9, 5))
plt.xlabel("Equipo")
plt.ylabel("Cantidad")
plt.title("Número de Victorias (W), Empates (D) y Derrotas (L) de los Equipos en La Liga 2023-24")
plt.xticks(range(len(data["Team"])), data["Team"], rotation=45)
plt.legend(["Victorias (W)", "Empates (D)", "Derrotas (L)"])
plt.show()
```

```python
# Gráfico de dispersión para ver puntos, goles a favor y goles en contra de los equipos.
plt.figure(figsize=(7, 4))
plt.scatter(data["GF"], data["GA"], c=data["Pts."], cmap="viridis")
plt.xlabel("Goles a Favor (GF)")
plt.ylabel("Goles en Contra (GA)")
plt.title("Relación entre Goles a Favor (GF) y Goles en Contra (GA) de los Equipos en La Liga 2023-24")
plt.colorbar(label="Puntos (Pts.)")
plt.show()
```

```python
# Cálculo de victorias, empates y derrotas.
total_victorias = data["W"].sum()
total_empates = data["D"].sum()
total_derrotas = data["L"].sum()

# Gráfico para observar la distribución de victorias, empates y derrotas de la temporada.
plt.figure(figsize=(6, 6))
labels = ["Victorias", "Empates", "Derrotas"]
sizes = [total_victorias, total_empates, total_derrotas]
colors = ["lightgreen", "lightblue", "lightcoral"]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.title("Distribución de Victorias, Empates y Derrotas en La Liga 2023-24")
plt.show()
```
